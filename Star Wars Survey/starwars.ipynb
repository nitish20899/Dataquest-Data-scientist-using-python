```python
import numpy as np
star_wars=star_wars[star_wars['RespondentID'].notnull()]
#remove null values from thr RespondentID column and the corresponding rows 
```

```python
star_wars.head()
```

```python
yes_no = {
    "Yes": True,
    "No": False
}
col=['Have you seen any of the 6 films in the Star Wars franchise?','Do you consider yourself to be a fan of the Star Wars film franchise?']
for i in col:
    star_wars[i]=star_wars[i].map(yes_no)
```

Do you consider yourself to be a fan of the Star Wars film franchise?

for the above columns we need to change the values from YES , NO to True and False.we do this to convert columns to Boolean type.

 columns should have intuitive names('seen_1', 'seen_2', 'seen_3', 'seen_4', 'seen_5', 'seen_6'), along with True and False values that indicate whether the respondent saw each of the six Star Wars movies.

so below we are going to change valuesto True and False and rename the columns

```python
star_wars.columns[3:9]
```

```python
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

```python
star_wars
```

```python
new_col=['seen_1','seen_2','seen_3','seen_4','seen_5','seen_6']
for i in range(6):
    star_wars=star_wars.rename(columns={(star_wars.columns[i+3]):new_col[i]})
```

```python
star_wars.columns[3:9]
```

```python
star_wars[star_wars.columns[9:15]]=star_wars[star_wars.columns[9:15]].astype(float)
#changing the ranking columns to float

newcol=['ranking_1','ranking_2','ranking_3','ranking_4','ranking_5','ranking_6']
for i in range(6):
    star_wars.rename(columns={star_wars.columns[i+9]:newcol[i]},inplace=True)
star_wars.columns[9:15]
#renaming the columns
```

we've cleaned up the ranking columns, we can find the highest-ranked movie more quickly. To do this, take the mean of each of the ranking columns using the pandas.DataFrame.mean() method on dataframes.

```python
import matplotlib.pyplot as plt
%matplotlib inline 
#importing matplotlib libraries

(star_wars[newcol].mean(axis=0)).plot(kind='bar')
#plot the bar graph with means of ranking 
```

As u can see ranking_5 is better (Remember that a lower ranking is better!)

Earlier in this project, we cleaned up the seen columns and converted their values to the Boolean type. When we call methods like pandas.DataFrame.sum() or mean(), they treat Booleans like integers. They consider True a 1, and False a 0. That means we can figure out how many people have seen each movie just by taking the sum of the column (even though they contain Boolean values)

```python
seen_sum=star_wars[star_wars.columns[3:9]].sum()
#number of people seen the  starwars episodes 
```

```python
seen_sum.plot(kind='bar')
```

As u can see episode 5 was mostly seen episode
